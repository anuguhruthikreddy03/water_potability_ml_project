```python
cols = df.columns
plt_n = 1
plt.figure(figsize=(18,10))
for i in cols:
    if i != "Potability":
        plt.subplot(3,3,plt_n)
        sns.boxplot(y=df[i],x=df['Potability'])
        m0 = df[df['Potability'] == 0][i].median()
        m1 = df[df['Potability'] == 1][i].median()
        plt.text(0, m0, f"{m0:.2f}", ha='center', va='bottom', fontweight='bold')
        plt.text(1, m1, f"{m1:.2f}", ha='center', va='bottom', fontweight='bold')
        plt.title(i)
        plt_n += 1
plt.tight_layout()
plt.show()
```

**Feature vs Potability for all features suggest that most overlap between potable and non-potable samples and with nealy identical spread adn meadians.This suggests limited discriminative power for a feature alone.Among all only solids show a modest shift in median and overlapping indicationg it is potentiall higher important feature**

```python
corr = df.drop(columns="Potability").corr()
plt.figure(figsize=(18,10))
sns.heatmap(corr,annot=True)
```

**Correlation analysis shows that no pair of features exhibits strong linear correlation , indicating an absence of multicollinearity. 
Therefore, all features were retained for subsequent modeling.**
